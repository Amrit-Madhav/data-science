## The Data

We will be working with a famous titanic data set for these exercises.

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style('whitegrid')
```

```python
titanic = sns.load_dataset('titanic')
```

```python
def pclass(x):
    if x == 1:
        return 'First'
    elif x == 2:
        return 'Second'
    else:
        return 'Third'

titanic['class'] = titanic['pclass'].apply(lambda x: pclass(x))
```

```python
def num2str(num):
    if num == 1:
        return "First"
    elif num == 2:
        return "Second"
    elif num == 3:
        return "Third"
    else:
        return "None"
    
cls = titanic['pclass'].apply(lambda x: num2str(x))
titanic['class1'] = cls
titanic.head()
import pandas as pd
import numpy as np
np.all(titanic['class']==titanic['class1'])
```

```python
with sns.axes_style(style='ticks'):
    g = sns.factorplot("class", "age", data=titanic, kind="box",order=["First", "Second","Third"])
    g.set_axis_labels("class", "age");
```

```python
ax = sns.swarmplot(x="class", y="age",
              palette="Set2", data=titanic,order=["First", "Second","Third"])
#ax.set_axis_labels("class", "age");
```

```python
ax = sns.countplot(x="sex", data=titanic,palette="BuGn")
```

```python
ax = sns.heatmap(titanic[["survived","pclass","age","sibsp","parch","fare","adult_male","alone"]].corr(), cmap="coolwarm_r",vmin=-1,vmax=1,annot=True, fmt=".2f")
```
